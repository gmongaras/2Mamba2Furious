```python
import torch
```

```python
N = 16
d = 8
e = 10
Q = torch.randn(N, d, dtype=torch.float64).cuda()
K = torch.randn_like(Q)
V = torch.randn(N, e, dtype=torch.float64).cuda()
```

```python
linear = (Q @ K.mT)**1 @ V
```

```python
linear = (Q @ K.mT)**1 @ V
linear_ = Q @ (K[:, :, None] * V[:, None, :]).sum(0)
torch.allclose(linear, linear_)
```

```python
quad = (Q @ K.mT)**2 @ V
quad_ = (Q[:, None, :] * Q[:, :, None]).flatten(-2, -1) @ (K[:, :, None, None] * K[:, None, :, None] * V[:, None, None, :]).sum(0).flatten(0, 1)
torch.allclose(quad, quad_)
```

```python
cube = (Q @ K.mT)**3 @ V
cube_ = (Q[:, :, None, None] * Q[:, None, :, None] * Q[:, None, None, :]).flatten(-3, -1) @ (K[:, :, None, None, None] * K[:, None, :, None, None] * K[:, None, None, :, None] * V[:, None, None, None, :]).sum(0).flatten(0, 2)
torch.allclose(cube, cube_)
```
